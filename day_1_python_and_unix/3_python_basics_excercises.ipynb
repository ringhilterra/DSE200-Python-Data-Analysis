(10) Write  a program that takes in a positive number (in some variable, say `i`) and computes the sum of all the number
between 0 and that number (inclusive). 

* Do it using a for loop
* Do it in one line using the function `sum` and list comprehension.

```python
# (10)

# 1 using a for loop
def sumNumLoop(i):
    sum = 0
    for num in range(1,i+1):
        sum += num
    return sum

# 2 using one line sum function list comprehension
def sumNumListComp(i):
    return sum([num for num in range(i+1)])
```

(13) Write a program to make a list of all prime numbers less than 3000. A prime number can only be divided evenly by 1 and itself.

```python
# (13)
def is_prime(n):
    if n >= 2:
        for x in range(2,n):
            if not ( n % x ):
                return False
    else:
        return False
    return True

# using list to keep track of prime numbers
primes = []
for i in range(3000):
    if is_prime(i):
        primes.append(i)
print(primes)
        
    
```
